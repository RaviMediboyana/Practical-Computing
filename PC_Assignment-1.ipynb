**Part-1** There are Six Faces for a cubiod as seen in the above Figure.But we have **Three Distinct Faces** So we will be using the Three Faces to Calculate  and Check which Face will be the Biggest Face in the cuboid.

**Part-2** We have to Calculate the Volume of the Rectangular cubiod.

```python
def Cuboid(l,b,h):
    First_Face = l*b
    Second_Face = b*h
    Third_Face = l*h
    
    biggest_face = max(First_Face,Second_Face,Third_Face)
    volume = l*b*h
    
    return biggest_face,volume
```

```python
# Calling the Cuboid Function and it will Return the biggest Face and the Volume Back.

Bigget_face,Volume = Cuboid(l=30,b=20,h=15)

print(f"The Biggest Face of the Cubiod is :{Bigget_face } cm\u00b2 and The Volume of the Cuboid is {Volume} cm\u00b3")
```

# 3-A) Unit Testing

>**Unit testing** involves the testing of each unit or an individual component of the software application. 
It is the first level of functional testing.The aim behind unit testing is to **validate unit components 
with its performance.**

>A unit is a single testable part of a software system and tested during the development phase of the application software.

>The purpose of unit testing is to test the correctness of isolated code. A unit component is an individual function or code of the application. **White box testing** approach used for unit testing and usually done by the developers.

>Whenever the application is ready and given to the Test engineer, he/she will start checking **every component of the module or module of the application independently or one by one**, and this process is known as Unit testing or components testing.

**Advantages**
>**1)** Unit testing uses module approach due to that any part can be tested without waiting for completion of another parts testing

>**2)** The developing team focuses on the provided functionality of the unit and how functionality should look in unit test suits to understand the unit API.

>**3)** Unit testing allows the developer to refactor code after a number of days and ensure the module still working without any defect.

**Disadvantages**
>**1)** It cannot identify integration or broad level error as it works on units of the code.

>**2)** In the unit testing, evaluation of all execution paths is not possible, so unit testing is not able to catch each and every error in a program.

>**3)** It is best suitable for conjunction with other testing activities.

```python
# The Required libraries are as follows.

import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
```

The Below code snippet is of Non_Vectorised method, we have created a funciton called as **Non_Vectorised** and we are passing the variable **n** where n is the size of the matrix and from this function we are returning the size of the matrix (i.e **n**) and the **time consumed** for calcualting the matrix.

```python
def Non_Vectorised(n):
    
    start = time.time() # this we will start the time as soon as it calls the function.
    
    mat1 = np.random.randint(10, size=(n, n)) # randomly creating the matrix of size n
    mat2 = np.random.randint(10, size=(n, n))
    
    result = np.zeros((n, n), dtype='int') # this is used to sort the values of the result.

    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            for k in range(len(mat2)):
                result[i][j] += mat1[i][k] * mat2[k][j]
                
    end = time.time() # As soon as the matrix result is created it will stop the time.
    
    Time_required = end-start # Here we will calculate the time required for the matrix to form.
    
    return Time_required ,n  # we are returning Time_required & the size of the matrix.
```

Here we are using the **Vectorised** method directly a funciton in numpy (i.e.**Matrix Multiplication ('np.matmul', 'np.dot' or '@'**) this all will give us the same result. 

```python
def Vectorize(n):
    start = time.time()
    mat1 = np.random.randint(1000, size=(n, n))
    mat2 = np.random.randint(1000, size=(n, n))
    
    product = np.matmul(mat1, mat2)
    end = time.time()
    Time_required = end-start
    return Time_required
```

In this question we want to check the time required for the Non_Vectorised method & The Vectorised Method. For this I feel we can use the **DataFrame** and we can compare the results both easily. and after that we will plot both the results.

```python
Data = pd.DataFrame() # Created a empty DataFrame to sort the time_reuqired and the size of the matrix.

Matrix_Size = [10,30,50,100,500,1000]  # this are the matrix sizes for which we have calculate the time.

for size in Matrix_Size:
    
    n = Non_Vectorised(size) # The Non_Vectorised Fucntion is called it will return the values and sore in n.
    N = Vectorize(size) # The Vectorised Fucntion is called.
    
    # We are append the values in the DataFrame.
    # Here we are Dividing by '60' to convert the time to 'mins' from 'seconds'
    
    Data = Data.append({'Size': n[1] ,'Time in min (Non_Vectorised)':(n[0]/60),'Time in min (Vectorised)':N/60},ignore_index=True)
```
